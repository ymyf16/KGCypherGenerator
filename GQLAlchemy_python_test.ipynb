```python
import pandas as pd
import numpy as np
from gqlalchemy import Memgraph
from gqlalchemy import match
from gqlalchemy.query_builders.memgraph_query_builder import Operator
```

```python
# Make a connection to the database
#memgraph = Memgraph(host='44.231.174.230', port=7687)
memgraph = Memgraph(host='alzkb.ai', port=7687)
```

```python
# Find Node Along with Its Count
query = memgraph.execute_and_fetch(
"""
MATCH (n) WITH LABELS(n) AS labels 
UNWIND labels AS label 
RETURN label, COUNT(*) AS count ORDER BY label;
""")
node_stats = pd.DataFrame(query)
```

```python
# Find Rel Along with Its Count
query = memgraph.execute_and_fetch(
"""
MATCH ()-[r]->()
RETURN type(r) AS RelationshipType, COUNT(r) AS NumberOfOccurrences;
""")
rel_stats = pd.DataFrame(query)
```
